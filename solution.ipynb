```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import re

# 1. Загрузка чистых данных заново
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

# 2. FEATURE ENGINEERING (Сборка всех лучших признаков)

# Обработка пропусков в Fare и Age медианой
test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].median())
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].median())

# Создание FamilySize и IsAlone
train_df["FamilySize"] = train_df['SibSp'] + train_df['Parch'] + 1
test_df["FamilySize"] = test_df['SibSp'] + test_df['Parch'] + 1
train_df['IsAlone'] = (train_df['FamilySize'] == 1).astype(int)
test_df['IsAlone'] = (test_df['FamilySize'] == 1).astype(int)

# 3. Подготовка данных для модели (Финальная версия)
features = ["Pclass", "Sex", "IsAlone", "Fare", "Age"] # Добавим Age!

# Преобразуем Sex в числа 0/1
train_df['Sex'] = train_df['Sex'].map({'female': 1, 'male': 0})
test_df['Sex'] = test_df['Sex'].map({'female': 1, 'male': 0})

X = train_df[features]
y = train_df["Survived"]
test_X = test_df[features]


# 4. Обучение модели Random Forest
model_rf_final = RandomForestClassifier(n_estimators=100, random_state=1)
model_rf_final.fit(X, y)

# 5. Предсказание и отправка
predictions_rf_final = model_rf_final.predict(test_X)
submission_rf_final_df = pd.DataFrame({'PassengerId': test_df['PassengerId'], 'Survived': predictions_rf_final})
submission_rf_final_df.to_csv('submission_random_forest_final.csv', index=False)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler # Импортируем StandardScaler

# Загрузка чистых данных (на всякий случай)
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

# Обработка пропусков (Age, Fare) и Feature Engineering (FamilySize, IsAlone)
test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].median())
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].median())
train_df["FamilySize"] = train_df['SibSp'] + train_df['Parch'] + 1
test_df["FamilySize"] = test_df['SibSp'] + test_df['Parch'] + 1
train_df['IsAlone'] = (train_df['FamilySize'] == 1).astype(int)
test_df['IsAlone'] = (test_df['FamilySize'] == 1).astype(int)

# Преобразование Sex в числа
train_df['Sex'] = train_df['Sex'].map({'female': 1, 'male': 0})
test_df['Sex'] = test_df['Sex'].map({'female': 1, 'male': 0})

# Выбор финальных признаков
features = ["Pclass", "Sex", "IsAlone", "Fare", "Age"]
X = train_df[features]
y = train_df["Survived"]
test_X = test_df[features]

# !!! МАШТАБИРОВАНИЕ ДАННЫХ !!!
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)         # Обучаем scaler и преобразуем обучающие данные
test_X_scaled = scaler.transform(test_X)  # Преобразуем тестовые данные (используя параметры из обучения)
```

```python
model_rf_scaled = RandomForestClassifier(n_estimators=100, random_state=1)
# Обучаем модель на МАШТАБИРОВАННЫХ данных
model_rf_scaled.fit(X_scaled, y) 

print("Модель Random Forest обучена на масштабированных данных!")
```

```python
# Делаем предсказание для МАШТАБИРОВАННЫХ тестовых данных
predictions_rf_scaled = model_rf_scaled.predict(test_X_scaled)

submission_rf_final_scaled_df = pd.DataFrame({'PassengerId': test_df['PassengerId'], 'Survived': predictions_rf_scaled})
submission_rf_final_scaled_df.to_csv('submission_random_forest_final_scaled.csv', index=False)
```

```python
# Делаем предсказание для МАШТАБИРОВАННЫХ тестовых данных
predictions_lr = model_lr.predict(test_X_scaled)

submission_lr_df = pd.DataFrame({'PassengerId': test_df['PassengerId'], 'Survived': predictions_lr})
submission_lr_df.to_csv('submission_logistic_regression.csv', index=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Определяем "сетку" параметров, которые мы хотим проверить
param_grid = {
    'n_estimators': [50, 100, 200],       # Количество деревьев
    'max_depth': [None, 5, 10, 15],       # Максимальная глубина деревьев
    'min_samples_split': [2, 5, 10],      # Минимальное количество выборок для разделения
    'min_samples_leaf': [1, 3, 5]         # Минимальное количество выборок в листе
}

# Создаем модель Random Forest
rf_model = RandomForestClassifier(random_state=1)

# Создаем объект GridSearchCV
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=5,            # Используем кросс-валидацию (5 разбиений данных)
    scoring='accuracy', # Метрика качества
    verbose=1,       # Выводим процесс обучения
    n_jobs=-1        # Используем все ядра процессора
)

# Запускаем поиск лучших параметров на масштабированных данных
grid_search.fit(X_scaled, y)

# Выводим лучшие найденные параметры и лучшую точность
print(f"\nЛучшие параметры: {grid_search.best_params_}")
print(f"Лучшая точность (кросс-валидация): {grid_search.best_score_}")
```
